# Importing libraries and loading dataset

```python
import pandas as pd

# Load the dataset from GitHub
url = 'https://raw.githubusercontent.com/aswintechguy/Machine-Learning-Projects/master/Wine%20Quality%20Prediction%20Analysis%20-%20Classification/winequality.csv'
wine_data = pd.read_csv(url)

# Display the first few rows of the dataset
wine_data
```

```python
# Display the summary statistics of the dataset
wine_data_description = wine_data.describe()
wine_data_description
```

```python
# Check for any missing values
missing_values = wine_data.isnull().sum()
missing_values
```

```python
# Display the distribution of the target variable 'quality'
quality_distribution = wine_data['quality'].value_counts()
quality_distribution
```

Twith the median) and to drop some columns which are not relevant.

```python
# Data Cleaning: Filling missing values with the median of each column
wine_data_cleaned = wine_data.fillna(wine_data.median())

# Confirming if all missing values are filled
missing_values_cleaned = wine_data_cleaned.isnull().sum()
missing_values_cleaned
```

# 1. Bottom Line Logistic Regression Model
- We start constructing an initial logistic regression model using all available predictors, including the 'type' of wine, which we will one-hot encode. We will use this Bottom Line Model to compare against future models.

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
import numpy as np


# One-hot encoding the 'type' column
wine_data_encoded = pd.get_dummies(wine_data_cleaned, columns=['type'])

# Separating features and target variable
x = wine_data_encoded.drop('quality', axis=1)
y = wine_data_encoded['quality']

# Splitting the dataset into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

# Scaling the features
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)

# Building the logistic regression model
model_baseline = LogisticRegression(max_iter=1000)
model_baseline.fit(x_train_scaled, y_train)

# Evaluating the baseline model
y_pred_baseline = model_baseline.predict(x_test_scaled)
bottom_line_accuracy = accuracy_score(y_test, y_pred_baseline)
print("Bottom Line Accuracy:", bottom_line_accuracy)
```

# 2. Data Preparation Phase: Transformation of the Predictors
- I decided to use the same predictor (Residual Sugar), and use different transformations on it, in order to observe the impact on the model performance.

## 2.1 Logarithmic Transformation on 'Residual Sugar'
- Applying a logarithmic transformation to 'Residual Sugar' to normalize its distribution.

```python
# Applying logarithmic transformation to 'Residual Sugar'
wine_data_encoded['residual sugar_log'] = np.log1p(wine_data_encoded['residual sugar'])

# Preparing the data with the transformed feature
x_log = wine_data_encoded.drop(['quality', 'residual sugar'], axis=1)
y_log = wine_data_encoded['quality']

# Splitting the data into training and testing sets
x_train_log, x_test_log, y_train_log, y_test_log = train_test_split(x_log, y_log, test_size=0.2, random_state=42)

# Scaling the transformed data
x_train_log_scaled = scaler.fit_transform(x_train_log)
x_test_log_scaled = scaler.transform(x_test_log)

# Building and evaluating the model
model_log = LogisticRegression(max_iter=1000)
model_log.fit(x_train_log_scaled, y_train_log)
y_pred_log = model_log.predict(x_test_log_scaled)
accuracy_log = accuracy_score(y_test_log, y_pred_log)
print("Accuracy with Logarithmic Transformation on 'Residual Sugar':", accuracy_log)
```

## 2.2 Square Root Transformation on 'Residual Sugar'
- The square root transformation is applied to 'Residual Sugar' to reduce skewness.

```python
# Applying square root transformation to 'Residual Sugar'
wine_data_encoded['residual sugar_sqrt'] = np.sqrt(wine_data_encoded['residual sugar'])

# Preparing the data with the transformed feature
x_sqrt = wine_data_encoded.drop(['quality', 'residual sugar', 'residual sugar_log'], axis=1)
y_sqrt = wine_data_encoded['quality']

# Splitting the data into training and testing sets
x_train_sqrt, x_test_sqrt, y_train_sqrt, y_test_sqrt = train_test_split(x_sqrt, y_sqrt, test_size=0.2, random_state=42)

# Scaling the transformed data
x_train_sqrt_scaled = scaler.fit_transform(x_train_sqrt)
x_test_sqrt_scaled = scaler.transform(x_test_sqrt)

# Building and evaluating the model
model_sqrt = LogisticRegression(max_iter=1000)
model_sqrt.fit(x_train_sqrt_scaled, y_train_sqrt)
y_pred_sqrt = model_sqrt.predict(x_test_sqrt_scaled)
accuracy_sqrt = accuracy_score(y_test_sqrt, y_pred_sqrt)
print("Accuracy with Square Root Transformation on 'Residual Sugar':", accuracy_sqrt)
```

## 2.3 Square Transformation on 'Residual Sugar'
- Applying a square transformation to 'Residual Sugar' to potentially reveal and leverage non-linear relationships with the target variable. This can be particularly useful for features that impact the target variable in a non-linear manner.

```python
# Applying square transformation to 'Residual Sugar'
wine_data_encoded['residual sugar_square'] = wine_data_encoded['residual sugar'] ** 2

x_square = wine_data_encoded.drop(['quality', 'residual sugar', 'residual sugar_log', 'residual sugar_sqrt'], axis=1)
y_square = wine_data_encoded['quality']

x_train_square, x_test_square, y_train_square, y_test_square = train_test_split(x_square, y_square, test_size=0.2, random_state=42)

x_train_square_scaled = scaler.fit_transform(x_train_square)
x_test_square_scaled = scaler.transform(x_test_square)

model_square = LogisticRegression(max_iter=1000)
model_square.fit(x_train_square_scaled, y_train_square)
y_pred_square = model_square.predict(x_test_square_scaled)
accuracy_square = accuracy_score(y_test_square, y_pred_square)
print("Accuracy with Square Transformation on 'Residual Sugar':", accuracy_square)
```

# 2.4 Reciprocal Transformation on 'Residual Sugar'
- The reciprocal transformation is applied to 'Residual Sugar' to explore its inverse relationship with the target variable.

```python
# Applying reciprocal transformation to 'Residual Sugar'
wine_data_encoded['residual sugar_reciprocal'] = 1 / (wine_data_encoded['residual sugar'] + 0.01)  # Added a small constant to avoid division by zero

x_reciprocal = wine_data_encoded.drop(['quality', 'residual sugar', 'residual sugar_log', 'residual sugar_sqrt', 'residual sugar_square'], axis=1)
y_reciprocal = wine_data_encoded['quality']

x_train_reciprocal, x_test_reciprocal, y_train_reciprocal, y_test_reciprocal = train_test_split(x_reciprocal, y_reciprocal, test_size=0.2, random_state=42)

x_train_reciprocal_scaled = scaler.fit_transform(x_train_reciprocal)
x_test_reciprocal_scaled = scaler.transform(x_test_reciprocal)

model_reciprocal = LogisticRegression(max_iter=1000)
model_reciprocal.fit(x_train_reciprocal_scaled, y_train_reciprocal)
y_pred_reciprocal = model_reciprocal.predict(x_test_reciprocal_scaled)
accuracy_reciprocal = accuracy_score(y_test_reciprocal, y_pred_reciprocal)
print("Accuracy with Reciprocal Transformation on 'Residual Sugar':", accuracy_reciprocal)
```

## 2.5 Box-Cox Transformation on 'Residual Sugar'
- Applying the Box-Cox transformation to 'Residual Sugar' to normalize its distribution.
- Note: This transformation method only works for positive data.

```python
from scipy import stats

# Ensure that 'Residual Sugar' has only positive values
wine_data_encoded['residual sugar'] += 1e-2  # Adding a small constant to avoid zero values

# Applying Box-Cox transformation to 'Residual Sugar'
wine_data_encoded['residual sugar_boxcox'], _ = stats.boxcox(wine_data_encoded['residual sugar'])

# Preparing the data with the transformed feature
x_boxcox = wine_data_encoded.drop(['quality', 'residual sugar', 'residual sugar_log', 'residual sugar_sqrt', 'residual sugar_square', 'residual sugar_reciprocal'], axis=1)
y_boxcox = wine_data_encoded['quality']

# Splitting the data into training and testing sets
x_train_boxcox, x_test_boxcox, y_train_boxcox, y_test_boxcox = train_test_split(x_boxcox, y_boxcox, test_size=0.2, random_state=42)

# Scaling the transformed data
x_train_boxcox_scaled = scaler.fit_transform(x_train_boxcox)
x_test_boxcox_scaled = scaler.transform(x_test_boxcox)

# Building and evaluating the model
model_boxcox = LogisticRegression(max_iter=1000)
model_boxcox.fit(x_train_boxcox_scaled, y_train_boxcox)
y_pred_boxcox = model_boxcox.predict(x_test_boxcox_scaled)
accuracy_boxcox = accuracy_score(y_test_boxcox, y_pred_boxcox)
print("Accuracy with Box-Cox Transformation on 'Residual Sugar':", accuracy_boxcox)
```

# 2.6 Power Transformation on 'Residual Sugar'
Applying a Power transformation to 'Residual Sugar' to improve normality. This transformation is useful for stabilizing variance.

```python
from sklearn.preprocessing import PowerTransformer

# Applying Power transformation to 'Residual Sugar'
power_transformer = PowerTransformer(method='yeo-johnson')
wine_data_encoded['residual sugar_power'] = power_transformer.fit_transform(wine_data_encoded[['residual sugar']])

# Preparing the data with the transformed feature
x_power = wine_data_encoded.drop(['quality', 'residual sugar', 'residual sugar_log', 'residual sugar_sqrt', 'residual sugar_square', 'residual sugar_reciprocal', 'residual sugar_boxcox'], axis=1)
y_power = wine_data_encoded['quality']

# Splitting the data into training and testing sets
x_train_power, x_test_power, y_train_power, y_test_power = train_test_split(x_power, y_power, test_size=0.2, random_state=42)

# Scaling the transformed data
x_train_power_scaled = scaler.fit_transform(x_train_power)
x_test_power_scaled = scaler.transform(x_test_power)

# Building and evaluating the model
model_power = LogisticRegression(max_iter=1000)
model_power.fit(x_train_power_scaled, y_train_power)
y_pred_power = model_power.predict(x_test_power_scaled)
accuracy_power = accuracy_score(y_test_power, y_pred_power)
print("Accuracy with Power Transformation on 'Residual Sugar':", accuracy_power)
```
